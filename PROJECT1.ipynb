```python
import getpass
import numpy as np

# Define a sample PIN (for demonstration purposes, in a real system this would be securely stored)
pin = getpass.getpass('Enter your 4-digit PIN: ')

# Define a transaction amount using numpy
transaction_amount = np.array(100.00, dtype=np.float64)

print(f"PIN entered: {'*' * len(pin)}")
print(f"Transaction Amount: ${transaction_amount:.2f}")
print(f"Type of transaction_amount: {type(transaction_amount)}")
```
